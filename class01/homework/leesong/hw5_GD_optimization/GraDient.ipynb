```python
import numpy as np
import matplotlib.pylab as plt
x = np.linspace(-2,2,11)
y = np.linspace(-2,2,11)

print(x)
print(y)

x,y = np.meshgrid(x,y)
print(x)
print(y)

f = lambda x,y : (x-1)**2 + (y-1)**2
z = f(x,y)
print(z)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
grad_f_x = lambda x, y: 2 * (x-1)
grad_f_y = lambda x, y: 2 * (y-1)

dz_dx = grad_f_x(x,y)
dz_dy = grad_f_y(x,y)

ax = plt.axes()
ax.contour(x,y,z, levels=np.linspace(0,10,20), cmap=plt.cm.jet)
ax.quiver(x,y,-dz_dx,-dz_dy)
ax.grid()
ax.axis('equal')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
plt.show()
```

```python
def f(x):
    return x**2 - 4*x + 6
```

```python
NumberOfPoints = 101
x = np.linspace(-5,5,NumberOfPoints)
fx = f(x)

plt.plot(x,fx)
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('plot of f(x)')
plt.show()
```

```python
xid = np.argmin(fx) #최소값 구하는 것
xopt = x[xid]
print(xopt, f(xopt))
```

```python
plt.plot(x,fx)
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('plot of f(x)')
plt.plot(xopt, f(xopt), 'xr') #최소값을 찍어줌
plt.show()
```

```python
def f(x):
    return x**2 - 4*x + 6

def grad_fx(x):
    return 2*x - 4

def steepest_descent(func, grad_func, x0, learning_rate=0.01, verbose=True, Maxiter=100):  # Maxiter 값이 설정되지 않아 추가하였습니다.
    paths = []
    for i in range(Maxiter):
        x1 = x0 - learning_rate * grad_func(x0)  # grad-fuc(x0) 부분을 grad_func(x0)로 수정하였습니다.
        if verbose:
            print('{0:03d} : {1:4.3f}, {2:4.2E}'.format(i, x1, func(x1)))
        x0 = x1
        paths.append(x0)
    return (x0, func(x0), paths)
```

```python
xopt, fopt, paths = steepest_descent(f, grad_fx, 0.0, learning_rate=1.2)  # 변수명이 잘못 입력되어 수정하였습니다.
```

```python
x = np.linspace(0.5, 2.5, 1000)
paths = np.array(paths)

plt.plot(x, f(x))  # plt.plt(x, f(x)) -> plt.plot(x, f(x))로 수정
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x)')
plt.plot(paths, f(paths), 'o-')  # plt.plot(paths, f(paths), 'o-')로 수정
plt.show()
```

```python
xopt, fopt, paths = steepest_descent(f, grad_fx, 1.0, learning_rate=1.0)  # 러닝 레이트 와 grad fx 값 변경가능.
```

```python
x = np.linspace(0.5, 3.5, 1000)
paths = np.array(paths)

plt.plot(x, f(x))  # plt.plt(x, f(x)) -> plt.plot(x, f(x))로 수정
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x)')
plt.plot(paths, f(paths), 'o-')  # plt.plot(paths, f(paths), 'o-')로 수정
plt.show()
```
